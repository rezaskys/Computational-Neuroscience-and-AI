# Drift diffusion model (docs)

## Theory

Why are we using a drift-diffusion model?



Going through the tutorial:

For a variable $x_k$ which is the state of the system which performs a 'random-walk'. The state at time 'k' is given by: $x_{k+1} = \lambda x_k$ where $\lambda$ is the 'rate of decay'. To make the process decay to a specific value of $x_\infty$, the new system now is: $x_{k+1} = x_\infty + \lambda (x_k - x_\infty)$. If the initial state of the system is $x_0$, we rewrite this as: $x_k = x_\infty(1 - \lambda^k) + x_0 \lambda^k$. To add a diffusion term, we have $x_{k+1} = x_\infty + \lambda (x_k - x_\infty) + \sigma \eta$ for $\eta \sim \mathcal{N}(\mu = 0, \sigma = 1)$

What if we are given $x$ as it evolves over time, and we want to recover the dynamics $\lambda$?

We think that $x_{k+1} = \lambda x_k + \eta$. Then, plotting $x_{k+1}$ against $x_k$ yields a plot which looks as if a regression line can be fit to the scatter data. Then, if we define $\mathbf{x}_1 = x_{0:T-1}, \mathbf{x}_2 = x_{1:T}$, we have $\mathbf{x}_2 = \lambda \mathbf{x}_1$ i.e. a regression model.

We can generalize this so that $x_k$ depends not only on $x_{k-1}$ but also previous terms. So, $x_{k+1} = \alpha_0 + \alpha_1 x_k + \alpha_2 x_{k-1} + ... + \alpha_{r+1} x_{k-r}$ for an $r$-order autoregressive model.

```python
# Read the dataset into a Pandas DataFrame
import pyddm
from pyddm import Sample
import pandas
with open(r'content/roitman_rts.csv' , "r") as f:
    df_rt = pandas.read_csv(f)

df_rt = df_rt[df_rt["monkey"] == 1] # Only monkey 1

# Remove short and long RTs, as in 10.1523/JNEUROSCI.4684-04.2005.
# This is not strictly necessary, but is performed here for
# compatibility with this study.
df_rt = df_rt[df_rt["rt"] > .1] # Remove trials less than 100ms
df_rt = df_rt[df_rt["rt"] < 1.65] # Remove trials greater than 1650ms
```

```python
roitman_sample = Sample.from_pandas_dataframe(df_rt, rt_column_name="rt", choice_column_name="correct")
```

> `roitman_sample = Sample.from_pandas_dataframe(df_rt, rt_column_name="rt", choice_column_name="correct")`

In the above code line, the choice_column_name is taken as "correct", where "correct" was in the original data as well. If our choice is -1, 0, 1 (0 being the no-response), I think we have the case for the boundaries corresponding to right and left, rather than correct and incorrect. We can have a second model for the correct vs incorrect case then.



```python
m = pyddm.gddm(drift=lambda coh, driftcoh : driftcoh*coh,
               noise=1,
               bound="b",
               nondecision="ndtime",
               parameters={"driftcoh": (-20,20), "b": (.4, 3), "ndtime": (0, .5)},
               conditions=["coh"])

# pyddm.plot.model_gui(m) # ...or...
pyddm.plot.model_gui_jupyter(m, sample=roitman_sample)
```

```python
# Fitting this will also be fast because PyDDM can automatically
# determine that DriftCoherence will allow an analytical solution.
m.fit(sample=roitman_sample, verbose=False) # Set verbose=True to see fitting progress
m.show()
print("Parameters:", m.parameters())
```

```python
# Plot the model fit to the PDFs and save the file.
import pyddm.plot
import matplotlib.pyplot as plt
pyddm.plot.plot_fit_diagnostics(model=m, sample=roitman_sample)
plt.savefig("roitman-fit.png")
plt.show()
```

#### Improving the fit

```python
model_leak = pyddm.gddm(
    drift=lambda driftcoh,leak,coh,x : driftcoh*coh - leak*x,
    bound=lambda bound_base,invtau,t : bound_base * np.exp(-t*invtau),
    nondecision="ndtime",
    parameters={"driftcoh": (-20,20),
                "leak": (-5, 5),
                "bound_base": (.5, 10),
                "ndtime": (0, .5),
                "invtau": (.1, 10)},
    conditions=["coh"])
```

*Issue with bottom cell:*

I'm facing an issue in computing `x_c` for c in {100, 50, 25, 12, 6, 0} in our data matrix `X`. The idea is that, it is the variable corresponding contrast value related to c shown in the particular trial `i` (regardless of left or right). For example, if `contrast[i]` = 0.0625, then `x_6[i]` will be +1, and -1 for other values of contrast.
So ideally, in matrix `X`, if we consider the columns for `x_c` (which are six columns in total, 1 for each contrast value), there should only be one +1 value in each row in these five columns (because only 1 value of contrast is there per trial) and all other entries among these columns should be -1.

```python
# From Paulami
def data_matrix_cond(non_neg):
    import numpy as np
    # from google.colab import files
    # Initialize arrays
    ones = np.ones(len(non_neg['contrastRight']))
    x_100 = np.zeros(len(non_neg['contrastRight']))
    x_25 = np.zeros(len(non_neg['contrastRight']))
    x_12 = np.zeros(len(non_neg['contrastRight']))
    x_6 = np.zeros(len(non_neg['contrastRight']))
    x_0 = np.zeros(len(non_neg['contrastRight']))
    x_side = np.zeros(len(non_neg['contrastRight']))
    x_f = np.zeros(len(non_neg['contrastRight']))
    x_block = non_neg['probabilityLeft']

    # Function to reset all contrast columns to -1
    def reset_contrast(i):
        x_100[i] = -1
        x_25[i] = -1
        x_12[i] = -1
        x_6[i] = -1
        x_0[i] = -1

    # Update feature arrays
    for i in range(len(non_neg['contrastRight'])):
        con = non_neg['contrastRight'][i]
        if i > 0:
            if non_neg['feedbackType'][i-1] == 1:
                x_f[i] = 1
            else:
                x_f[i] = -1

        if np.isnan(con):
            x_side[i] = -1
            reset_contrast(i)
        else:
            x_side[i] = 1
            reset_contrast(i)
            if con == 1.0:
                x_100[i] = 1
            elif con == 0.25:
                x_25[i] = 1
            elif con == 0.125:
                x_12[i] = 1
            elif con == 0.0625:
                x_6[i] = 1
            else:
                x_0[i] = 1

    for i in range(len(non_neg['contrastLeft'])):
        con = non_neg['contrastLeft'][i]
        if not np.isnan(con):
            reset_contrast(i)
            if con == 1.0:
                x_100[i] = 1
            elif con == 0.25:
                x_25[i] = 1
            elif con == 0.125:
                x_12[i] = 1
            elif con == 0.0625:
                x_6[i] = 1
            else:
                x_0[i] = 1

    # Construct design matrix
    X = ones
    for col in [x_100, x_25, x_12, x_6, x_0, x_side, x_f, x_block]:
        X = np.column_stack((X, col))
    return X

```
